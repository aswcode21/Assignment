```python
# Scatter Plot:

# Use: Scatter plots are used to visualize the relationship between two continuous variables.
# Example: You can use a scatter plot to examine the correlation between a person's age and their income.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
tips = sns.load_dataset("tips")

# Create scatter plot
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.show()
```

```python
# Line Plot:

# Use: Line plots are commonly used to display trends over time.
# Example: You can create a line plot to track the daily closing prices of a stock over a period of several months.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
fmri = sns.load_dataset("fmri")

# Create line plot
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event")
plt.show()
```

```python
# Bar Plot:

# Use: Bar plots are used to compare categorical data or display the frequency distribution of a categorical variable.
# Example: You can create a bar plot to compare the sales performance of different products in a store.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
titanic = sns.load_dataset("titanic")

# Create bar plot
sns.barplot(data=titanic, x="class", y="fare", hue="survived")
plt.show()
```

```python
# Histogram:

# Use: Histograms are used to visualize the distribution of a single continuous variable.
# Example: You can create a histogram to analyze the distribution of student scores in a class.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
tips = sns.load_dataset("tips")

# Create histogram
sns.histplot(data=tips, x="total_bill", kde=True)
plt.show()
```

```python
# Box Plot:

# Use: Box plots provide a summary of the distribution of a continuous variable and help identify potential outliers.
# Example: You can create a box plot to compare the distribution of exam scores across different schools.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
iris = sns.load_dataset("iris")

# Create box plot
sns.boxplot(data=iris, x="species", y="sepal_length")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load "titanic" dataset
titanic = sns.load_dataset("titanic")

# Create box plots for age and fare based on passenger class
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title("Box Plot of Age by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")

plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title("Box Plot of Fare by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")

plt.tight_layout()
plt.show()
```

```python
# plt.subplot(1, 2, 1) and plt.subplot(1, 2, 2):

# The subplot function is used to create a grid of subplots within a single figure.
# In this case, we want to create two subplots side by side, so we specify (1, 2) as the grid shape.
# The third argument, 1 and 2, specifies the position of each subplot in the grid.
# By using plt.subplot(1, 2, 1) and plt.subplot(1, 2, 2), we are creating two subplots side by side.
# plt.tight_layout():

# The tight_layout function is used to automatically adjust the subplot parameters to improve the spacing between subplots and labels.
# When multiple subplots are created, there can be overlapping or insufficient space between them.
# By calling plt.tight_layout(), we ensure that the subplots and their labels are properly spaced and do not overlap, providing a more
# visually appealing and readable plot.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load "diamonds" dataset
diamonds = sns.load_dataset("diamonds")

# Create histogram with hue parameter
sns.histplot(data=diamonds, x='price', hue='cut', kde=True)
plt.title("Histogram of Diamond Prices with Cut")
plt.xlabel("Price")
plt.ylabel("Count")
plt.legend(title="Cut")
plt.show()
```
