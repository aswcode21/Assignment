```python
# Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with
# an example.
```

```python
# Probability Mass Function (PMF):
# The Probability Mass Function (PMF) is applicable to discrete random variables. 
# It gives the probability of each possible outcome in the sample space of the random variable. 
# In other words, it provides the probability that the random variable takes on a specific value.
# For a discrete random variable X, the PMF is denoted as P(X = x), where "x" represents a specific value in the sample space of X.
# The PMF satisfies the following properties:

# P(X = x) ≥ 0 for all x in the sample space of X.
# The sum of all probabilities in the PMF is equal to 1.

# Example of PMF:
# Let's consider the random variable X representing the outcome of rolling a fair six-sided die.
# The sample space of X is {1, 2, 3, 4, 5, 6}. Since the die is fair, each outcome has an equal probability of 1/6.

# The PMF for this example is:
# P(X = 1) = 1/6
# P(X = 2) = 1/6
# P(X = 3) = 1/6
# P(X = 4) = 1/6
# P(X = 5) = 1/6
# P(X = 6) = 1/6

def fair_die_pmf(outcome):
    if outcome in range(1, 7):
        return 1/6
    else:
        return 0

# Calculate the PMF for each outcome
outcomes = [1, 2, 3, 4, 5, 6]
pmf_values = [fair_die_pmf(outcome) for outcome in outcomes]

# Print the PMF values
for outcome, pmf in zip(outcomes, pmf_values):
    print(f'P(X = {outcome}) = {pmf}')
```

```python
# Probability Density Function (PDF):
# The Probability Density Function (PDF) is applicable to continuous random variables.
# Unlike the PMF, the PDF does not provide the exact probability of a specific outcome. 
# Instead, it gives the relative likelihood of the random variable falling within a particular interval.
# The area under the PDF curve within a specific interval corresponds to the probability of the random variable falling in that interval.
# For a continuous random variable X, the PDF is denoted as f(x). The PDF satisfies the following properties:

# f(x) ≥ 0 for all x.
# The total area under the PDF curve is equal to 1.
# Example of PDF:
# Let's consider a continuous random variable X representing the height of adult males. 
# The PDF for this variable might be a bell-shaped curve, such as the normal distribution.

# In this example, we can't assign specific probabilities to individual heights,
# but we can determine the probability of a random adult male having a height within a certain range.

# For instance, if the PDF of adult male height follows a normal distribution with a mean of 175 cm and a 
# standard deviation of 6 cm, we can calculate the probability of a male being between 170 cm and 180 cm tall 
# by finding the area under the curve within that interval.

# These concepts, PMF and PDF, are fundamental in understanding and analyzing the behavior of random variables
# and their associated probabilities in both discrete and continuous cases.
import numpy as np
import matplotlib.pyplot as plt

def normal_pdf(x, mean, std_dev):
    return (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)

# Generate data for the PDF curve
mean_height = 175
std_dev_height = 6
x_values = np.linspace(mean_height - 3 * std_dev_height, mean_height + 3 * std_dev_height, 1000)
pdf_values = normal_pdf(x_values, mean_height, std_dev_height)

# Plot the PDF curve
plt.plot(x_values, pdf_values, label='PDF')
plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')
plt.title('Normal Distribution - Probability Density Function')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?
```

```python
# he Cumulative Density Function (CDF) is a concept used in probability theory and statistics to describe 
# the cumulative probability distribution of a random variable.
# It gives the probability that the random variable takes on a value less than or equal to a specific value.
# In other words, the CDF provides a measure of the likelihood that the random variable is less than or equal to a certain value.

# For a random variable X, the CDF is denoted as F(x), and it is defined as follows:
# F(x) = P(X ≤ x)

# The CDF has several important properties:

# F(x) is a non-decreasing function, i.e., as x increases, the cumulative probability also increases or remains the same.
# The value of F(x) lies between 0 and 1, inclusive, for all x.
```

```python
def fair_die_pmf(outcome):
    if outcome in range(1, 7):
        return 1/6
    else:
        return 0

def fair_die_cdf(x):
    cumulative_prob = 0
    for outcome in range(1, x + 1):
        cumulative_prob += fair_die_pmf(outcome)
    return cumulative_prob

# Calculate the CDF for each outcome
outcomes = [1, 2, 3, 4, 5, 6]
cdf_values = [fair_die_cdf(outcome) for outcome in outcomes]

# Print the CDF values
for outcome, cdf in zip(outcomes, cdf_values):
    print(f'F(X ≤ {outcome}) = {cdf}')
```

```python
# The Cumulative Density Function (CDF) is used for various reasons:

# Determining probabilities: The CDF allows us to determine the probability that a random variable
# falls within a certain range of values. For example, the probability of a value between a and b can be calculated as F(b) - F(a).

# Quantiles and percentiles: CDF enables us to find quantiles, which are specific points in the distribution
# that divide it into equal or predetermined proportions. For example, the median is the 50th percentile, and the first quartile is the 25th percentile.

# Comparing distributions: CDFs provide an easy way to compare different probability distributions and understand 
# their characteristics, such as location, spread, and shape.

# Simulation and modeling: CDFs are used in various statistical models and simulations, such as Monte Carlo simulations, 
# to generate random samples from specific distributions.
```

```python
# The discrete probability distribution of a random variable which takes value 1 with probability p and the value 0 with probability q=1-p. 
# Less probably,It can be thought of as a model for set of possible outcomes of any  single experiment that asks a yes-no question.
```

```python
# The probability mass function (PMF) of the Bernoulli distribution is as follows:

# P(X = 1) = p
# P(X = 0) = 1 - p

# Where:

# P(X = 1) is the probability of success (getting a 1),
# P(X = 0) is the probability of failure (getting a 0),
# p is the probability of success in a single trial, and
# 1 - p is the probability of failure in a single trial.

# Example:
# Let's consider a simple example of flipping a fair coin. In this case, the outcome of each flip can either
# be "Heads" (H) or "Tails" (T), making it a binary outcome. Let's assume that getting "Heads" is considered a success (1), 
# and getting "Tails" is considered a failure (0).

# Here, the probability of success (getting "Heads") is p = 0.5, and the probability of failure (getting "Tails") is 1 - p = 0.5.

# The Bernoulli distribution for this example can be represented as follows:
# P(X = 1) = 0.5 (Probability of getting "Heads")
# P(X = 0) = 0.5 (Probability of getting "Tails")


# Difference between Bernoulli Distribution and Binomial Distribution:

# Definition:
# Bernoulli Distribution: Models a single binary outcome (success or failure) in a single trial.
# Binomial Distribution: Models the number of successes in a fixed number of independent Bernoulli trials.
# Parameters:
# Bernoulli Distribution: Has a single parameter "p," which represents the probability of success in a single trial.
# Binomial Distribution: Has two parameters: "n" (the number of trials) and "p" (the probability of success in a single trial).
# Number of Trials:
# Bernoulli Distribution: Considers only one trial, i.e., a single event with two possible outcomes.
# Binomial Distribution: Considers multiple independent trials with the same probability of success in each trial.
# Outcome:
# Bernoulli Distribution: Provides the probability of a specific outcome (success or failure) in a single trial.
# Binomial Distribution: Gives the probability of obtaining a specific number of successes in "n" independent Bernoulli trials.
# Probability Mass Function (PMF):
# Bernoulli Distribution: The PMF is P(X = 1) = p and P(X = 0) = 1 - p.
# Binomial Distribution: The PMF gives the probability of getting exactly "k" successes in "n" trials and is
# represented by the formula P(X = k) = C(n, k) * p^k * (1 - p)^(n - k), where C(n, k) is the binomial coefficient ("n choose k").
```

```python
# Q7: Explain uniform Distribution with an example.
```

```python
# continous uniform distribution

# In probability theory and statistics, the continous uniform distribution or rectanglular distribution is a family of symmetric 
# probability distributions. The distribution describes an experiment where there is an arbitary outcome that lies between certain bounds.
# The bounds are defined by by parameters,a and b,which are minimum and maximum values.

# example-Let's consider an example of a continuous uniform distribution over the interval [2, 8]. 
# In this case, any value between 2 and 8 (inclusive) is equally likely to occur.

# The probability density function (PDF) of the continuous uniform distribution for this example is:

# f(x) = 1 / (8 - 2) = 1/6 for 2 ≤ x ≤ 8
# f(x) = 0 otherwise

# Here, any value within the interval [2, 8] has a constant probability density of 1/6, 
# indicating that each value in this range is equally likely to occur.

# Visually, the PDF of the continuous uniform distribution over the interval [2, 8] would appear as a
# flat horizontal line at a height of 1/6, indicating that the probability of any value between 2 and 8 is the same.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the interval [a, b]
a = 10
b = 20

# Generate random numbers from the continuous uniform distribution
num_samples = 1000
random_values = np.random.uniform(a, b, num_samples)

# Plot a histogram to visualize the distribution of random values
plt.hist(random_values, bins=20, density=True, alpha=0.6, color='g')
plt.xlabel('Random Values')
plt.ylabel('Probability Density')
plt.title('Continuous Uniform Distribution')
plt.grid(True)
plt.show()
```

```python
# Discrete uniform distribution
# the discrete uniform distribution is a symmetric probability distribution where a finite number of values 
# are equally likely to be observed,every one of n values has equal probability 1/n.

# Let's consider an example of rolling a fair six-sided die. In this case, the possible outcomes of rolling 
# the die are the integers from 1 to 6. Each outcome is equally likely to occur, making it a discrete uniform distribution.

# For this example, the set of possible values is {1, 2, 3, 4, 5, 6}, and the size of the set (N) is 6.

# The probability mass function (PMF) of the discrete uniform distribution for this example is:

# P(X = 1) = 1/6
# P(X = 2) = 1/6
# P(X = 3) = 1/6
# P(X = 4) = 1/6
# P(X = 5) = 1/6
# P(X = 6) = 1/6

# Here, each outcome has a probability of 1/6, indicating that each face of the fair die is equally likely to occur when rolled.
```

```python
import numpy as np

# Define the possible outcomes of the four-sided die
possible_outcomes = [1, 2, 3, 4]

# Number of rolls to simulate
num_rolls = 100

# Simulate the rolls of the die
rolls = np.random.choice(possible_outcomes, num_rolls)

# Count the occurrences of each outcome
outcome_counts = {outcome: np.count_nonzero(rolls == outcome) for outcome in possible_outcomes}

# Calculate the probability of each outcome
total_rolls = len(rolls)
outcome_probabilities = {outcome: count / total_rolls for outcome, count in outcome_counts.items()}

# Print the results
print("Simulated Die Rolls:")
print(rolls)
print("\nOutcome Counts:")
print(outcome_counts)
print("\nOutcome Probabilities:")
print(outcome_probabilities)
```

```python
# Q8: What is the z score? State the importance of the z score.
```

```python
# The Z-score, also known as the standard score or standardized score, is a statistical measure that quantifies
# how many standard deviations a data point is from the mean of a dataset. It indicates whether a particular data
# point is above or below the mean and how far it deviates from the mean in terms of standard deviations.

# The formula to calculate the Z-score for a data point "x" in a dataset with mean "μ" and standard deviation "σ" is:

# Z = (x - μ) / σ


# The Z-score allows us to standardize data and compare values from different datasets that may have different units or scales.
# It is particularly important in the field of statistics and data analysis for several reasons:

# Standardization: The Z-score transforms raw data into a standardized scale with a mean of 0 and a standard deviation of 1. 
# This standardization makes it easier to compare and interpret data, as it eliminates the effect of the original units or scales.

# Outlier Detection: Z-scores help identify outliers in a dataset. Data points with Z-scores significantly 
# higher or lower than 0 (usually above 2 or below -2) are considered outliers and might require further investigation.

# Probability Calculations: The Z-score is used in various statistical calculations and hypothesis testing. 
# It allows us to find the probability of a data point falling within a specific range, given the mean and standard deviation of the dataset.

# Normal Distribution: In a standard normal distribution (mean = 0, standard deviation = 1), 
# the Z-score represents the position of a data point relative to the mean. Positive Z-scores indicate data points 
# above the mean, while negative Z-scores indicate data points below the mean.

# Comparing Data Points: Z-scores enable direct comparison of data points from different datasets. 
# This is particularly useful when comparing measurements or observations from different populations.

# Data Transformation: Z-scores are used in various data transformations and normalization techniques in machine learning and statistical modeling.
```

```python
 #Q9.What is Central Limit Theorem? State the significance of the Central Limit Theorem.
```

```python
# The Central Limit Theorem (CLT) is a fundamental concept in statistics that states that the sum or
# average of a large number of independent, identically distributed random variables will tend to follow a
# normal distribution, regardless of the underlying distribution of the original variables. 
# In simpler terms, the sampling distribution of the sample mean approaches a normal distribution as the 
# sample size increases, even if the individual data points do not follow a normal distribution.

# The Central Limit Theorem is significant for several reasons:

# Normality Assumption: The CLT allows us to approximate the sampling distribution of the mean for a population, 
# even when the population distribution is not normal. This is particularly useful in cases where the population 
# distribution is unknown or not easily identifiable.

# Basis for Inference: The CLT is the foundation for many statistical inference methods, such as hypothesis 
# testing and confidence intervals. These methods rely on the assumption of a normal distribution, which can often be approximated using the CLT.

# Sample Size Determination: The CLT helps in determining the sample size required for statistical analyses. 
# As the sample size increases, the distribution of the sample mean becomes more normal, leading to more reliable statistical inferences.

# Aids in Decision-Making: The CLT allows us to make statistical decisions and draw conclusions about a population based on a sample. 
# It provides a way to understand the behavior of sample statistics and their relationship to the population parameters.

# Data Analysis: In practical data analysis, it is often assumed that the sampling distribution of the sample mean or
# sum is approximately normal, allowing for the use of various statistical methods that rely on normality.

# Generalization: The CLT is widely applicable to a broad range of random variables and distributions, 
# making it a universal tool in statistics and data analysis.

# Importance in Quality Control: In quality control and process improvement, the CLT is essential for analyzing the 
# variability of processes and ensuring that products meet specific quality standards.
```
